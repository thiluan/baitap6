```python
import turtle
import math
#thiết lập màn hình và bút vẽ
sc=turtle.Screen()
sc.bgcolor("sky blue")
t=turtle.Turtle()
t.pensize(5)
t.pencolor("black")
t.fillcolor("yellow")
# Thiết lập kích thước cạnh tường nhà
w=200
h=100
# Vẽ tường nhà
t.begin_fill()
for i in range(2):
    t.forward(w)
    t.left(90)
    t.forward(100)
    t.left(90)
t.end_fill()    
```

```python
#thiết lập kích thước cửa chính
wd=50
hd=70
#vẽ cửa chính
t.goto(75,0)
t.fillcolor("orange")
t.begin_fill()
for i in range(2):
    t.forward(wd)
    t.left(90)
    t.forward(hd)
    t.left(90)
t.end_fill()    
```

```python
# vẽ mái nhà
t.penup()
t.goto(140,100)
t.fillcolor("red")
t.pendown()
t.begin_fill()
t.left(120)
t.forward(140)
t.left(120)
t.forward(140)
t.left(120)
t.forward(140)
t.end_fill()
```

```python
# vẽ ống khói
t.penup()
t.left(120)
t.forward(90)
t.right(30)
t.pendown()
t.begin_fill()
t.fillcolor("brown")
t.forward(50)
t.left(90)
t.forward(20)
t.left(90)
t.forward(15)
t.left(30)
t.forward(35)
t.end_fill()
            
```

```python
# vẽ khói
t.penup()
t.goto(90,260)
t.left(150)
t.pendown()
t.forward(50)
t.penup()
t.goto(75,275)
t.pendown()
t.forward(30)
```

```python
# vẽ thân cây
t.penup()
t.goto(-100,-100)
t.pendown()
t.begin_fill()
t.forward(100)
t.left(90)
t.forward(40)
t.left(80)
t.forward(100)
t.left(100)
t.forward(60)
t.end_fill()
# vẽ lá cây 1
t.penup()
t.goto(-120,0)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.forward(50)
t.left(120)
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)
t.forward(50)
t.end_fill()
# vẽ lá cây 2
t.penup()
t.goto(-120,75)
t.pendown()
t.begin_fill()
t.forward(50)
t.left(120)
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)
t.forward(50)
t.end_fill()
#vẽ lá cây 3
t.penup()
t.goto(-120,150)
t.pendown()
t.begin_fill()
t.forward(50)
t.left(120)
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)
t.forward(50)
t.end_fill()
```
